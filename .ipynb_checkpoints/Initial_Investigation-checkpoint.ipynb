```python
mu = btc["Close Price"].mean()
std = btc["Close Price"].std()
```

Notes for creating a small auto-trading model:


```python
close_prices = np.random.normal(mu, std, 100)
plt.plot(close_prices)
plt.title('BTC Normally-distributed random close prices')
plt.ylabel('BTC Price')
plt.xlabel('Time Point')
```

```python

def run_simulation(prices):
    
    catalog = {}
    times_events = {}
    mean_z_scores = []
    for i, p in enumerate(prices):
        if i >= 6:
            tzero = i - 6
            t_one = i - 5
            current_data = prices[t_one:i]
            z_score = scipy.stats.zscore(current_data)
            mean_z_scores.append(z_score.mean())
            if current_data.mean() > current_data[0] and z_score.mean() > 0.0:
                times_events[i] = 'sell'
                if "sell" not in catalog:
                    catalog["sell"] = 1
                else:
                    catalog["sell"] += 1
            else:
                times_events[i] = 'buy'
                if "buy" not in catalog:
                    catalog["buy"] = 1
                else:
                    catalog["buy"] += 1
                    
    return catalog, times_events, np.asarray(mean_z_scores)

catalog, all_events, mean_z = run_simulation(close_prices)
```

```python

catalog, all_events, mean_z = run_simulation(close_prices)

def create_df(events_dict):
    dex = ["Event Type"]
    events = pd.DataFrame(events_dict, dex).T
    events["Categorize"] = float("Nan")
    events["Categorize"][events["Event Type"] == 'buy'] = 1
    events["Categorize"][events["Event Type"] == 'sell'] = 0
    return events

events = create_df(all_events)
binary = np.asarray(events["Categorize"])
time = np.asarray(events.index)
print catalog
plt.scatter(time, binary)
#plt.plot(mean_z)
```
